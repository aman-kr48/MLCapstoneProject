## 4. DATA CLEANING AND PREPARATION

```python
ctc = strat_train_set.drop("Expected_CTC", axis=1)
ctc_labels = strat_train_set["Expected_CTC"].copy()
```

#### Feature Engineering

```python
def preprocess(data):

    # Add Job_Experience
    data["Job_Experience"] = (data["Industry"] != "Missing").astype(int)

    Education_year = ["Passing_Year_Of_Graduation", "Passing_Year_Of_PG", "Passing_Year_Of_PHD"]
    data["Education_Completion_Year"] = data[Education_year].max(axis=1) # Add Education_Completion_Year
    data["Education_Completion_Year"].fillna(2025, inplace=True) # Fill NAN with 2025
    data["Career_Start_Year"] = data["Education_Completion_Year"]- data["Total_Experience"] # Add Career_Start_Year
    data.drop("Education_Completion_Year", axis=1, inplace=True) # Drop Education_Completion_Year

    # Map Education and create a new column Education_Level
    education_map = {
        'Under Grad': 1,
        'Grad': 2,
        'PG': 3,
        'Doctorate': 4}
    
    data["Education_Level"] = data["Education"].map(education_map)

    # Map Inhand_offer and create a new column Offer
    offer_map = {'N': 0, 'Y': 1}
    
    data["Offer"] = data["Inhand_Offer"].map(offer_map)

    # Add Overall_Profile
    data["Overall_Profile"] = data["Education_Level"] + data["Offer"]

    # Drop Education and Inhand_Offer    
    data.drop("Education", axis=1, inplace=True)
    data.drop("Inhand_Offer", axis=1, inplace=True)


    # Drop all the remaining less important columns
    drop_columns = ['IDX', 'Applicant_ID', 'Industry','Organization', 'Graduation_Specialization', 'University_Grad',
                   'Passing_Year_Of_Graduation', 'PG_Specialization', 'University_PG', 'Passing_Year_Of_PG', 
                    'PHD_Specialization', 'University_PHD', 'Passing_Year_Of_PHD', 'Curent_Location', 'Preferred_location']
    
    data.drop(drop_columns, axis=1, inplace=True)
```

```python
preprocess(ctc)
```

#### Pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
ctc.info()
```

```python
ctc["Last_Appraisal_Rating"].value_counts()
```

```python
custom_order = [["Missing", 'D', 'C', 'B', 'A', 'Key_Performer']]
```

```python
ordinal_pipeline = Pipeline([('imputer', SimpleImputer(strategy= "constant", fill_value= "Missing")),
                            ('encoder', OrdinalEncoder(categories=custom_order))])
```

```python
cat_pipeline = Pipeline([('imputer', SimpleImputer(strategy= "constant", fill_value= "Missing")),
                            ('encoder', OneHotEncoder())])
```

```python
num_pipeline = Pipeline([('std_scaler', StandardScaler())])
```

```python
ord_attribs = ["Last_Appraisal_Rating"]
cat_attribs = ["Department", "Role", "Designation"]
columns = ctc.columns.tolist()
```

```python
num_attribs = [x for x in columns if x not in ord_attribs + cat_attribs]
```

```python
full_pipeline = ColumnTransformer([('num', num_pipeline, num_attribs),
                         ('ord', ordinal_pipeline, ord_attribs),
                         ('cat', cat_pipeline, cat_attribs)])
```

```python
ctc_prepared = full_pipeline.fit_transform(ctc)
```

## 5. TRAIN AND SELECT A MODEL

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(ctc_prepared, ctc_labels)
```

```python
from sklearn.metrics import mean_squared_error
ctc_predictions = lin_reg.predict(ctc_prepared)
lin_mse = mean_squared_error(ctc_labels, ctc_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse 
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(ctc_prepared, ctc_labels)
```

```python
ctc_predictions = tree_reg.predict(ctc_prepared)
tree_mse = mean_squared_error(ctc_labels, ctc_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, ctc_prepared, ctc_labels, scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)   
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
display_scores(tree_rmse_scores)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor(random_state=42)
forest_reg.fit(ctc_prepared, ctc_labels)
```

```python
ctc_predictions = forest_reg.predict(ctc_prepared)
forest_mse = mean_squared_error(ctc_labels, ctc_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(forest_reg, ctc_prepared, ctc_labels, scoring="neg_mean_squared_error", cv=5)
forest_rmse_scores = np.sqrt(-scores)
```

```python
display_scores(forest_rmse_scores)
```

## 6. FINE-TUNE THE SELECTED MODEL

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_dist = {
    'n_estimators': randint(50, 150),    
    'max_depth': randint(10, 40),            
    'max_features': randint(5, 20),         
    'min_samples_split': randint(2, 10),     
    'min_samples_leaf': randint(1, 5),       
}
```

```python
random_search = RandomizedSearchCV(
    RandomForestRegressor(random_state=42),
    param_distributions=param_dist,
    n_iter=20,                
    cv=3,                     
    scoring='neg_mean_squared_error',
    return_train_score=True,
    n_jobs=-1)
```

```python
random_search.fit(ctc_prepared, ctc_labels)
```

```python
random_search.best_params_
```

```python
random_search.best_estimator_
```

```python
cvres = random_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

#### In this capstone project, I performed a regression task using RandomForestRegressor. I performed two randomised searches and one grid search to tune the model. But the regressor with the default setting performed better than the fine-tuned models.umn.

#### Careful analysis of the categorical columns using groupby plots showed that many categorical columns were useless for CTC prediction, so I dropped them in the preprocessing step.

#### Finally, some new features like Overall_Profile and Career_Start_Year were highly correlated with Expected_CTC. The Overall_Profile column was added because for freshers it is the biggest predictor of the expected CTC according to my analysis. The Education feature was replaced with Education_Level, but Ordinal encoding could have been used instead. I also added Job_Experience just to compensate for dropping the Industry column.
